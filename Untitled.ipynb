```python
from PyPDF2 import PdfFileReader

pdfFileObj = open('data_files/Win32_Industroyer.pdf', 'rb')
pdfReader = PdfFileReader(pdfFileObj)
```

```python
pdfReader.numPages
```

```python
#     text = pageObj.extractText()
#     print(text)    

key = '/Annots'
uri = '/URI'
ank = '/A'

for num in range(0,17):
    pageSliced  = pdfReader.getPage(num)
    pageObject = pageSliced.getObject()

    if pageObject.has_key(key):
        ann = pageObject[key]
        for a in ann:
            u = a.getObject()
            if u[ank].has_key(uri):
                print (u[ank][uri], "p{}".format(num))
```
